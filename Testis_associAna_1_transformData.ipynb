生成anndata的数据

每一个睾丸样本为一行，基因名存于列之中。用基因缩写，不用Ensembl Gene ID。


```python
def combineSASP(SA,SP):
    ANN_SP = SP
    ANN_SA = SA    

    myANN = ANN_SA
    for i in ANN_SA.index:
        sampID = ANN_SA.loc[i,"SAMPID"]
        for j in ANN_SP.index:
            subjID = ANN_SP.loc[j,"SUBJID"]
            if subjID in sampID:
                mySex = ANN_SP.loc[j,"SEX"]
                myAge = ANN_SP.loc[j,"AGE"]
                myDT = ANN_SP.loc[j,"DTHHRDY"]
                myANN.loc[i, 'SUBJID'] = subjID
                myANN.loc[i, 'SEX'] = mySex
                myANN.loc[i, 'AGE'] = myAge
                myANN.loc[i, 'DTHHRDY'] = myDT

    GTEx_v8_ANN = myANN
    return GTEx_v8_ANN
```

```python

def getTissueData(myTissue = "Testis"): 
    myGTEx_Data_FH = gzip.open('GTEx_Analysis_v10_RNASeQCv2.4.2_gene_tpm.gct.gz', 'rt', encoding='utf-8')
    myGTEx_SA = pd.read_csv("GTEx_Analysis_v10_Annotations_SampleAttributesDS.txt",sep="\t")
    myGTEx_SP = pd.read_csv("GTEx_Analysis_v10_Annotations_SubjectPhenotypesDS.txt",sep="\t")
    myGTEx_ANN = combineSASP(myGTEx_SA,myGTEx_SP)
    #myGTEx_Tissue = myGTEx_ANN[myGTEx_ANN.SMTS == myTissue]
    myGTEx_Tissue = myGTEx_ANN[myGTEx_ANN.SMTSD == myTissue]
    myGTEx_Tissue_SAMPID = myGTEx_Tissue.SAMPID.to_list()
    
    l = myGTEx_Data_FH.readline()
    mySAMPID_index = []
    mySAMPID = []
    ################################
    while l != "":
        ls = l.strip().split("\t")
        if len(ls) > 1:
            if ls[0] == "Name":
                for i in ls:
                    if i in myGTEx_Tissue_SAMPID:
                        mySAMPID_index.append(ls.index(i))
                        mySAMPID.append(i)
                break
        l = myGTEx_Data_FH.readline()
    
    myGTEx_dic = {}
    for id in mySAMPID:
        myGTEx_dic[id] =[]
    
    myEGIs = []
    mySymbols = []
    
    l = myGTEx_Data_FH.readline()
    while l != "":
        ls = l.strip().split("\t")
        myEGIs.append(ls[0])
        mySymbols.append(ls[1])
        
        myValues = np.array(ls)[mySAMPID_index]
        myValues = myValues.tolist()
        myDic = dict(zip(mySAMPID,myValues))
        for id in mySAMPID:
            myGTEx_dic[id].append(myDic[id])
        
        l = myGTEx_Data_FH.readline()
    
    myGTEx_Data_FH.close()
    
    
    myGenes = dict(zip(myEGIs,mySymbols))
    
    myPD = pd.DataFrame(myGTEx_dic)
    myPD.index = myEGIs
    myAnnData = ad.AnnData(myPD.T.to_numpy())
    myAnnData.obs_names = myPD.columns
    myAnnData.var_names = myPD.index
    
    ###########################################
    mySMPTHNTS_dic = dict(zip(myGTEx_ANN.SAMPID,myGTEx_ANN.SMPTHNTS))
    mySMPTHNTS = []
    for id in myPD.columns:
        mySMPTHNTS.append(mySMPTHNTS_dic[id])
    
    myAnnData.obs["SMPTHNTS"] = mySMPTHNTS
    
    #########
    mySMTSD_dic = dict(zip(myGTEx_ANN.SAMPID,myGTEx_ANN.SMTSD))
    mySMTSD = []
    for id in myPD.columns:
        mySMTSD.append(mySMTSD_dic[id])
    
    myAnnData.obs["SMTSD"] = pd.Categorical(mySMTSD)
    
    #########
    mySUBJID_dic = dict(zip(myGTEx_ANN.SAMPID,myGTEx_ANN.SUBJID))
    mySUBJID = []
    for id in myPD.columns:
        mySUBJID.append(mySUBJID_dic[id])
    myAnnData.obs["SUBJID"] = mySUBJID
    
    #########
    mySEX_dic = dict(zip(myGTEx_ANN.SAMPID,myGTEx_ANN.SEX))
    mySEX = []
    for id in myPD.columns:
        mySEX.append(mySEX_dic[id])
    
    myAnnData.obs["SEX"] = mySEX
    
    #########
    myAGE_dic = dict(zip(myGTEx_ANN.SAMPID,myGTEx_ANN.AGE))
    myAGE = []
    for id in myPD.columns:
        myAGE.append(myAGE_dic[id])
    
    myAnnData.obs["AGE"] = myAGE
    
    #########
    myDTHHRDY_dic = dict(zip(myGTEx_ANN.SAMPID,myGTEx_ANN.DTHHRDY))
    myDTHHRDY = []
    for id in myPD.columns:
        myDTHHRDY.append(myDTHHRDY_dic[id])
    
    myAnnData.obs["DTHHRDY"] = myDTHHRDY
    
   
    ##############################
    #myGenes = dict(zip(myEGIs,mySymbols))
    mySymbols = []
    for id in myPD.index:
        mySymbols.append(myGenes[id])
    myAnnData.var["Symbol"] = mySymbols
    
    
    return myAnnData

```
